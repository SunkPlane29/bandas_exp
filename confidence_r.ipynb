```r
df <- read.table("EoS-insensitive_posterior_samples.dat", header=FALSE, col.names=c("m1", "m2", "lambda1", "lambda2", "r1", "r2"), skip=1)
```

```r
df
```

```r
r = as.numeric(unlist(c(df["m2"])))
m = as.numeric(unlist(c(df["lambda2"])))
```

```r
library("r2d2")
```

```r
plot(r, m)
surf <- MASS::kde2d(r, m, n=19)
region <- conf2d_int(r, m, surf, level=0.90, n=300)
polygon(region, lwd=2)
```

```r
wdf <- data.frame(region[1], region[2])
colnames(wdf) <- c("ConfRegionM", "ConfRegionLambda")
write.csv(wdf, "GW170817_EoS_insensitive_mlambda90conf_2.
```
