```r
df <- read.table("EoS-insensitive_posterior_samples.dat", header=FALSE, col.names=c("m1", "m2", "lambda1", "lambda2", "r1", "r2"), skip=1)

x = as.numeric(unlist(c(df["r1"])))
y = as.numeric(unlist(c(df["m1"])))

plot(x, y)
surf <- MASS::kde2d(x, y, n=300)
region <- conf2d_int(x, y, surf, level=0.90, n=300)
polygon(region, lwd=2)

wdf <- data.frame(region[1], region[2])
colnames(wdf) <- c("ConfRegionR", "ConfRegionM")
write.csv(wdf, "GW170817_EoS_insensitive_mr90conf_1.csv", row.names=FALSE)
```
